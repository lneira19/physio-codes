```python
# Libraries

import numpy as np
import matplotlib.pyplot as plt
import random
import matplotlib.animation as animation
```

```python
c_min = 20
c_max = 1000
step = 10

# 0-    Generation of columns with each amount of items.

arr = np.arange(start=c_min,stop=c_max+step,step=step)
arr = arr[::-1]

print(arr)

# 1-    How many items each column will lose?

lost_items = np.array([np.uint32(random.randint(a=0,b=item)) for item in arr])

print(lost_items)

# 2-    Having determined the lost items for each column,
#       how many of those will go to the column to the left and to the right?  

lost_items_division = []
for i in lost_items:
    left = np.uint32(random.randint(a=0,b=i))
    right = i - left
    lost_items_division.append([left,right])
lost_items_division = np.array(lost_items_division)

print(lost_items_division)

# 3-    Getting the new values of the columns
new_arr = []
for i in range(len(arr)):
    if i == 0:
        new_arr.append(np.uint32(c_max))
        continue
    elif i == len(arr)-1:
        new_arr.append(np.uint32(c_min))
    else:
        new_item = arr[i] - lost_items[i] + lost_items_division[i-1][1] + lost_items_division[i+1][0]
        new_arr.append(np.uint32(new_item))
new_arr = np.array(new_arr)

print(new_arr)
```

```python
def redistribution(arr,
                   c_max:np.uint32  = 1000,
                   c_min:np.uint32  = 20):
    
    # 1-    How many items each column will lose?
    
    lost_items = np.array([np.uint32(random.randint(a=0,b=item)) for item in arr])

    # 2-    Having determined the lost items for each column,
    #       how many of those will go to the column to the left and to the right?  

    lost_items_division = []
    for i in lost_items:
        left = np.uint32(random.randint(a=0,b=i))
        right = i - left
        lost_items_division.append([left,right])
    lost_items_division = np.array(lost_items_division)

    # 3-    Getting the new values of the columns
    
    new_arr = []
    for i in range(len(arr)):
        if i == 0:
            new_arr.append(np.uint32(c_max))
            continue
        elif i == len(arr)-1:
            new_arr.append(np.uint32(c_min))
        else:
            new_item = arr[i] - lost_items[i] + lost_items_division[i-1][1] + lost_items_division[i+1][0]
            new_arr.append(np.uint32(new_item))
    new_arr = np.array(new_arr)

    return new_arr
```

```python
c_min = 20
c_max = 1000
step = 10
iterations = 200

n = np.arange(start=1,stop=len(arr)+1)
arr = np.arange(start=c_min,stop=c_max+step,step=step)
arr = arr[::-1]


fig, ax = plt.subplots()

fig.set_figheight(5)
fig.set_figwidth(12)

ax.set_xlabel("Column")
ax.set_ylabel("Concentration")

ax.set_title("Redistribution of particles")

ax.grid(linestyle="--")


ims = []
for i in range(iterations):

    im3 = ax.text(x=np.uint32(0.8*len(arr)),y=2*c_max,s=f"Frame nÂº{i}")
    
    if i == 0:
        new_arr = arr
        im1, = ax.plot(n,arr,color="green")
        im2 = ax.scatter(n,arr,color="red")
        ims.append([im1,im2,im3])
        continue
    else:
        new_arr = redistribution(new_arr,c_max=c_max,c_min=c_min)
        im1, = ax.plot(n,new_arr,color="green")
        im2 = ax.scatter(n,new_arr,color="red")
        ims.append([im1,im2,im3])

ani = animation.ArtistAnimation(fig, ims, interval=1000, blit=True, repeat_delay=1000)
writer = animation.PillowWriter(fps=15, metadata=dict(artist='lneira19'),bitrate=1800)
ani.save('flow.gif', writer=writer)

plt.close()
```
